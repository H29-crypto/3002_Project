```python
import os
import torch
import torchvision
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim

# Path to the cropped & sorted data
data_dir = '/content/drive/MyDrive/3002/ClassifierData'

# Data augmentation & normalization
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
])

# Datasets and loaders
dataset = datasets.ImageFolder(data_dir, transform=transform)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

# Label mapping
class_names = dataset.classes
print("Detected Classes:", class_names)

# Model setup (ResNet18)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet18(pretrained=True)
model.fc = nn.Linear(model.fc.in_features, len(class_names))  # 5 classes
model = model.to(device)

# Training settings
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
epochs = 10

# Training loop
for epoch in range(epochs):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for inputs, labels in loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = torch.argmax(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss:.4f}, Accuracy: {correct/total:.2f}")

# Save the model
torch.save(model.state_dict(), "/content/drive/MyDrive/3002/threat_classifier.pt")
```

```python
from ultralytics import YOLO
import torch
import torchvision.transforms as transforms
from torchvision import models
import cv2
import os
from PIL import Image
import numpy as np

# Load YOLOv12 model
yolo_model = YOLO("/content/drive/MyDrive/3002/yolo_model_object_detection/weights/best.pt")

# Load ResNet18 classifier
classifier = models.resnet18(weights='IMAGENET1K_V1')
classifier.fc = torch.nn.Linear(classifier.fc.in_features, 5)
classifier.load_state_dict(torch.load("/content/drive/MyDrive/3002/threat_classifier.pt", map_location='cpu'))
classifier.eval()

# Class labels
class_names = ['Gun', 'Knife', 'Pliers', 'Scissors', 'Wrench']

# Transforms for classifier input
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# Path to test images
test_path = "/content/drive/MyDrive/3002/Data/test/images"
output_folder = "/content/drive/MyDrive/3002/double_check_results"
os.makedirs(output_folder, exist_ok=True)

for image_name in os.listdir(test_path):
    if not image_name.endswith((".jpg", ".png")):
        continue

    img_path = os.path.join(test_path, image_name)
    img_cv2 = cv2.imread(img_path)
    results = yolo_model(img_path)[0]

    for i, box in enumerate(results.boxes):
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        crop = img_cv2[y1:y2, x1:x2]

        # Convert crop to PIL for classifier
        crop_pil = Image.fromarray(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
        input_tensor = transform(crop_pil).unsqueeze(0)  # Add batch dimension

        # Get prediction
        with torch.no_grad():
            logits = classifier(input_tensor)
            pred_class = class_names[logits.argmax().item()]
            confidence = torch.softmax(logits, dim=1).max().item()

        # Annotate image
        cv2.rectangle(img_cv2, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img_cv2, f'{pred_class} ({confidence:.2f})', (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Save result
    out_path = os.path.join(output_folder, image_name)
    cv2.imwrite(out_path, img_cv2)

print("YOLO + Classifier pipeline completed and results saved.")
```

```python
import os
import torch
from torchvision import models, transforms
from PIL import Image
import numpy as np
import cv2
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image

# Paths
crops_folder = "/content/drive/MyDrive/3002/Crops1"
output_folder = "/content/drive/MyDrive/3002/GradCAM_Results"
os.makedirs(output_folder, exist_ok=True)

# Load classifier
classifier = models.resnet18(pretrained=False)
classifier.fc = torch.nn.Linear(classifier.fc.in_features, 5)  # for 5 classes
state_dict = torch.load("/content/drive/MyDrive/3002/threat_classifier.pt", map_location="cpu")
classifier.load_state_dict(state_dict)
classifier.eval()

# Grad-CAM setup
target_layers = [classifier.layer4[-1]]
cam = GradCAM(model=classifier, target_layers=target_layers)

# Transform
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Process each image
for img_file in os.listdir(crops_folder):
    if not img_file.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    img_path = os.path.join(crops_folder, img_file)
    image = Image.open(img_path).convert("RGB")
    rgb_img = np.array(image.resize((224, 224))) / 255.0
    input_tensor = transform(image).unsqueeze(0)

    # Optional: predict class
    with torch.no_grad():
        outputs = classifier(input_tensor)
        pred_class = torch.argmax(outputs, dim=1).item()

    grayscale_cam = cam(input_tensor=input_tensor, targets=[ClassifierOutputTarget(pred_class)])[0]
    cam_image = show_cam_on_image(rgb_img, grayscale_cam, use_rgb=True)

    # Save the Grad-CAM result
    save_path = os.path.join(output_folder, f"gradcam_{img_file}")
    cv2.imwrite(save_path, cv2.cvtColor(cam_image, cv2.COLOR_RGB2BGR))

print(" All Grad-CAM visualizations saved!")
```

```python
import os
import torch
import cv2
from torchvision import transforms, models
from ultralytics import YOLO
from PIL import Image
import numpy as np

# Paths
YOLO_MODEL_PATH = "/content/drive/MyDrive/3002/yolo_model_object_detection/weights/best.pt"
CLASSIFIER_PATH = "/content/drive/MyDrive/3002/threat_classifier.pt"
INPUT_IMAGES_PATH = "/content/drive/MyDrive/3002/Data/test/images"
OUTPUT_FOLDER = "/content/drive/MyDrive/3002/Filtered_Results"
CONFIDENCE_THRESHOLD = 0.7  # for classifier

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# Load YOLOv12 model
yolo_model = YOLO(YOLO_MODEL_PATH)

# Load classifier model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
classifier = models.resnet18(weights="DEFAULT")
classifier.fc = torch.nn.Linear(classifier.fc.in_features, 5)  # 5 classes
classifier.load_state_dict(torch.load(CLASSIFIER_PATH, map_location=device))
classifier = classifier.to(device)
classifier.eval()

# Preprocessing for classifier
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

class_names = ['Gun', 'Knife', 'Pliers', 'Scissors', 'Wrench']

# Run YOLO and filter with classifier
for filename in os.listdir(INPUT_IMAGES_PATH):
    if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
        image_path = os.path.join(INPUT_IMAGES_PATH, filename)
        original_img = cv2.imread(image_path)

        results = yolo_model(image_path)[0]

        for i, box in enumerate(results.boxes):
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            crop = original_img[y1:y2, x1:x2]

            # Convert to PIL and preprocess
            pil_crop = Image.fromarray(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
            input_tensor = transform(pil_crop).unsqueeze(0).to(device)

            # Classify
            with torch.no_grad():
                outputs = classifier(input_tensor)
                prob = torch.softmax(outputs, dim=1)
                confidence, predicted = torch.max(prob, 1)
                confidence = confidence.item()
                predicted_class = class_names[predicted.item()]

            # Accept only high-confidence predictions
            if confidence >= CONFIDENCE_THRESHOLD:
                save_path = os.path.join(OUTPUT_FOLDER, f"{predicted_class}_{confidence:.2f}_{filename}")
                cv2.imwrite(save_path, crop)

print("✅ Filtered detections saved successfully.")
```

```python
from PIL import Image
import numpy as np
import torch
from torchvision import transforms
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from torchvision.transforms.functional import to_pil_image

# Load image
image = Image.open(image_path).convert("RGB")
image_np = np.array(image)

# Run YOLO
results = yolo_model(image_path)[0]

# Set up transform
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Loop over detections
for i, box in enumerate(results.boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    crop = image_np[y1:y2, x1:x2]
    crop_tensor = transform(crop).unsqueeze(0)

    # Classify crop
    with torch.no_grad():
        pred = classifier(crop_tensor)
        prob = torch.softmax(pred, dim=1)
        conf, pred_cls = torch.max(prob, 1)
        label = CLASSES[pred_cls]

    print(f"Detection {i+1}: {label} ({conf.item():.2f})")

    # Grad-CAM setup
    cam = GradCAM(model=classifier, target_layers=[classifier.layer4[-1]])

    # Grad-CAM target class
    targets = [ClassifierOutputTarget(pred_cls.item())]
    grayscale_cam = cam(input_tensor=crop_tensor, targets=targets)[0]

    # Convert crop_tensor to image and overlay Grad-CAM
    input_image_np = np.array(to_pil_image(crop_tensor.squeeze(0))).astype(np.float32) / 255.0
    cam_image = show_cam_on_image(input_image_np, grayscale_cam, use_rgb=True)

    # Show both images
    Image.fromarray(crop).show()
    Image.fromarray(cam_image).show()
```
